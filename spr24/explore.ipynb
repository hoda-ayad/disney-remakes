# Movie Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
links = pd.read_csv('arc_links.csv')
nodes = pd.read_csv('arc_nodes.csv')
```

```python
cols = nodes.columns
nodes = nodes.rename(columns={cols[0]: 'item', cols[1]:'link', cols[2]:'image link', cols[3]: 'type', cols[4]:'year', cols[5]:'endyear', cols[6]:'og', cols[7]:'adapt', cols[8]:'life?'})
```

```python
box_off = nodes[nodes['year'] == 2023]

plt.pie(box_off['adapt'].value_counts(), labels=box_off['adapt'].value_counts().keys())
```
